```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the first 5 dataset
df_tracks = pd.read_csv('spotify_data.csv')
df_tracks.head()
```

```python
#To check null values in the dataset
pd.isnull(df_tracks).sum()
```

```python
# Data Inspection and details
df_tracks.info()
```

```python
#Top 10 list popularity songs
sorted_df = df_tracks.sort_values('track_popularity', ascending = True).head(10)
sorted_df
```

```python
#to get descriptive types of the dataset
df_tracks.describe().transpose()
```

```python
#10 most popular songs>90
most_track_popular = df_tracks.query('track_popularity>90',inplace = False).sort_values('track_popularity', ascending = False)
most_track_popular[:10]
```

```python
#To load data
df = pd.read_csv('spotify_data.csv')
df
```

```python
#To filter out the particular data from dataset
df_tracks[['track_artist']].iloc[10]
```

```python
#To check the duration of the dataset
df_tracks.duration.head()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "loudness", x="energy", color="c").set(title="Loudness vs Energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "track_popularity", x="acousticness", color="b").set(title="Track_Popularity vs acousticness correlation")
```

```python
df_genre=pd.read_csv("spotify_data.csv")
```

```python
df_genre.head()
```

```python
plt.title("Duratin of the songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='playlist_genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style= "darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("track_popularity", ascending = False).head(10)
sns.barplot(y='playlist_genre',x='track_popularity', data = famous).set(title= "Top 5 genres by popularity")
```

```python
# --- Data Visualization 
# 1. Distribution of Danceability
plt.figure(figsize=(10, 6))
sns.histplot(df['danceability'], bins=20, kde=True, color='purple')
plt.title('Distribution of Song Danceability', fontsize=16)
plt.xlabel('Danceability Score', fontsize=12)
plt.ylabel('Number of Tracks', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('danceability_distribution.png')
plt.close()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(df['danceability'], bins=20, kde=True, color='purple')
plt.title('Distribution of Song Danceability', fontsize=16)
plt.xlabel('Danceability Score', fontsize=12)
plt.ylabel('Number of Tracks', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('danceability_distribution.png')
plt.close()
```

```python
sns.barplot(x='track_popularity', y='track_name', data=sorted_df, palette='viridis', hue='track_name', legend=False)
plt.xlabel('Popularity Score', fontsize=12)
plt.ylabel('Track Name', fontsize=12)
plt.title('Top 10 Most Popular Tracks', fontsize=16)
plt.tight_layout()
plt.savefig('top_10_popular_tracks.png')
plt.close()
```
